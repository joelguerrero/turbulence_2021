```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Remember to erase last parentehses when reading in the input fiels

#RE=100000
#y+ less than 1
data1 = np.loadtxt('l10t_yp1',skiprows=4)

#1 < y+ < 6
data2 = np.loadtxt('l10t_yp2',skiprows=4)

#y+ > 30
data3 = np.loadtxt('l10t_yp3',skiprows=4)

#y+ > 30 no inflation layer
data4 = np.loadtxt('l10t_yp4',skiprows=4)


#RE=100000 compressible
#y+ less than 1
data1comp = np.loadtxt('l10_yp1_compressible',skiprows=4)
```

```python
U=1
rho=1

#y+ less than 1
mu1=1.e-6
nu1=mu1/rho
#ws1=0.00226424
ws1=0.00226137

#1 < y+ < 6
mu2=1.e-6
nu2=mu1/rho
#ws2=0.00227262
ws2=0.00226645

#y+ > 30
mu3=1.e-6
nu3=mu1/rho
#ws3=0.00220864
ws3=0.00219378

#y+ > 30 no inflation layer
mu4=1.e-6
nu4=mu1/rho
#ws4=0.00231377
ws4=0.00231154


mid1=int(len(data1)/2)
mid2=int(len(data2)/2)
mid3=int(len(data3)/2)
mid4=int(len(data4)/2)


utau1=np.sqrt(ws1/rho)
#ypn1=utau1*(data1[0:mid1+1,0]+0.05)/nu1
ypn1=utau1*(abs(data1[0:mid1+1,0]-data1[0,0]))/nu1
upn1=data1[0:mid1+1,1]/utau1

utau2=np.sqrt(ws2/rho)
ypn2=utau2*(abs(data2[0:mid2+1,0]-data2[0,0]))/nu2
upn2=data2[0:mid2+1,1]/utau2

utau3=np.sqrt(ws3/rho)
ypn3=utau3*(abs(data3[0:mid3+1,0]-data3[0,0]))/nu3
upn3=data3[0:mid3+1,1]/utau3

utau4=np.sqrt(ws4/rho)
ypn4=utau4*(abs(data4[0:mid4+1,0]-data4[0,0]))/nu4
upn4=data4[0:mid4+1,1]/utau4
```

```python
#Compressible case
Uc=15.21
rhoc=1.176655

#y+ less than 1
mu1c=1.7894e-05
nu1c=mu1c/rhoc
#ws1=0.00226424
ws1c=0.666766

mid1c=int(len(data1comp)/2)

utau1c=np.sqrt(ws1c/rhoc)
#ypn1=utau1*(data1[0:mid1+1,0]+0.05)/nu1
ypn1c=utau1c*(abs(data1comp[0:mid1c+1,0]-data1[0,0]))/nu1c
upn1c=data1comp[0:mid1c+1,1]/utau1c
```

```python
#Compute therorical profiles
x=np.linspace(0.1,100000,1000000)
yp=x

#Laminar
upl=yp
#Low-law
upt=(1.0/0.41)*np.log(yp)+5.0

#Compute therorical profiles
x1=np.linspace(0.1,20,10000)
x2=np.linspace(5,100000,10000)

yp1=x1
yp2=x2

#Laminar
up1=yp1

#Low-law
up2=(1.0/0.41)*np.log(yp2)+5.0
```

```python
#Spalding universal law

up3 = np.linspace(0,30,10000)

#E = 9.025
#k = 0.4
#B = 5.5

E = 9.1
k = 0.42

#E = 8.6
#k = 0.417
    
sl = up3 + (1/E)*( np.exp(k*up3) - 1 - (k*up3) - ((k*up3)**2)/2  - ((k*up3)**3)/6 - ((k*up3)**4)/24 )
```
