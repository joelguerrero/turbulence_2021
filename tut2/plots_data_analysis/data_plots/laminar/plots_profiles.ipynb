```python
U=1
rho=1

#y+ less than 1
mu1=0.0002
nu1=mu1/rho
ws1=0.0158916

#1 < y+ < 6
mu2=0.0002
nu2=mu1/rho
ws2=0.0159003

#y+ > 30
mu3=0.0002
nu3=mu1/rho
ws3=0.0158907

#y+ > 30 no inflation layer
mu4=0.0002
nu4=mu1/rho
ws4=0.0159026


mid1=int(len(data1)/2)
mid2=int(len(data2)/2)
mid3=int(len(data3)/2)
mid4=int(len(data4)/2)


utau1=np.sqrt(ws1/rho)
#ypn1=utau1*(data1[0:mid1+1,0]+0.05)/nu1
ypn1=utau1*(abs(data1[0:mid1+1,0]-data1[0,0]))/nu1
upn1=data1[0:mid1+1,1]/utau1

utau2=np.sqrt(ws2/rho)
ypn2=utau2*(abs(data2[0:mid2+1,0]-data2[0,0]))/nu2
upn2=data2[0:mid2+1,1]/utau2

utau3=np.sqrt(ws3/rho)
ypn3=utau3*(abs(data3[0:mid3+1,0]-data3[0,0]))/nu3
upn3=data3[0:mid3+1,1]/utau3

utau4=np.sqrt(ws4/rho)
ypn4=utau4*(abs(data4[0:mid4+1,0]-data4[0,0]))/nu4
upn4=data4[0:mid4+1,1]/utau4
```

```python
#Compute therorical profiles
x=np.linspace(0.1,100000,1000000)
yp=x

#Laminar
up=yp
#Low-law
up1=(1.0/0.41)*np.log(yp)+5.0
```

```python
#Spalding universal law

up3 = np.linspace(0,30,10000)

#E = 9.025
#k = 0.4
#B = 5.5

E = 9.1
k = 0.42

#E = 8.6
#k = 0.417
    
sl = up3 + (1/E)*( np.exp(k*up3) - 1 - (k*up3) - ((k*up3)**2)/2  - ((k*up3)**3)/6 - ((k*up3)**4)/24 )
```

```python
#Plot profiles

plt.figure(figsize=(14, 8))

#Correlations
plt.plot(yp,up,label='Viscous sublayer - $U^+=y^+$')
plt.plot(yp,up1,label='Log-law - $U^+ = \dfrac{1}{\kappa} ln y^+ + C^+$')

plt.plot(sl,up3,'o-',label='Spalding Law of the Wall',markerfacecolor='white',markevery=300,zorder=10)

#FLUENT
plt.plot(ypn1[1:],upn1[1:],'-o',ms=6,label='Fluent - $y^+ < 1$',markerfacecolor='white',zorder=11)
plt.plot(ypn2[1:],upn2[1:],'-o',ms=6,label='Fluent - 1 < $y^+ < 6$',markerfacecolor='white',zorder=12)
plt.plot(ypn3[1:],upn3[1:],'-o',ms=6,label='Fluent $y^+ > 30$',markerfacecolor='white',zorder=13)
plt.plot(ypn4[1:],upn4[1:],'-o',ms=6,label='Fluent $y^+ > 30$ - No BL layer',markerfacecolor='white',zorder=14)

plt.xscale('log')
#plt.xlim(0,10000)
plt.ylim(0,40)

plt.grid()

plt.tick_params(labelsize=16)
plt.legend(loc=0,fontsize=16)

plt.xlabel('$y^+$', fontsize=20)
plt.ylabel('$U^+$', fontsize=20)

plt.tight_layout()

#plt.savefig("fig5.png")
```
