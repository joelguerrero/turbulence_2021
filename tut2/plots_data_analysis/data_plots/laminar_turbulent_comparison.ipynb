```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#RE=500

#Mesh 1 - Super fine at the wall
data1 = np.loadtxt('laminar/tm_off/yp1/l10',skiprows=4)

#Mesh 1 - Super fine at the wall
data2 = np.loadtxt('turbulent/l10t_yp1',skiprows=4)
```

```python
#Laminar Re = 500
#U=1
#rho=1
#mu1=0.0002
#nu1=mu1/rho

#Turbulent Re = 100000
#U=1
#rho=1
#mu1=0.000001
#nu1=mu1/rho
```

```python
#Plot profiles

plt.figure(figsize=(14, 8))

#Fluent
plt.plot(data1[:,1],data1[:,0],'-o',ms=6,label='Fluent - Mesh 1 - Re = 500',markerfacecolor='white',zorder=8)
plt.plot(data2[:,1],data2[:,0],'-o',ms=6,label='Fluent - Mesh 1 - Re = 100000',markerfacecolor='white',zorder=9)


x=np.linspace(0.0,0.05,100)
plt.plot (1.15242*(1-x/0.05)**(1./8),x+0.0,label= "Tubulent correlation - Power law",color='green')
plt.plot (1.15242*(1-x/0.05)**(1./8),-x+0.0,color='green')

x=np.linspace(-0.05,0.05,100)
plt.plot (max(data1[:,1])*(1-x**2/0.05**2),x,label= "Laminar correlation - Hagen-Poiseuille solution", color="red")

#plt.xscale('log')
#plt.xlim(0,10000)
plt.ylim(-0.06,0.06,10)

plt.grid()

plt.tick_params(labelsize=16)
plt.legend(loc=0,fontsize=16)

plt.xlabel('$U$', fontsize=20)
plt.ylabel('$y$', fontsize=20)

plt.tight_layout()

#plt.savefig("fig1.png")
#plt.savefig("fig1.pdf")
```
